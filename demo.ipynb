```python
print('We''ll do some math ...\n')
x = int(input('Enter an integer: '))
print('Powers of', x,'from 0 to 10:')
for i in range(11):
    print(f'{i:4} {x**i}')

print('\nRoots of', x,'from 1 to 10:')
for i in range(1,11):
    print(f'{i:4} {x**(1/i):12.5f}')

print('\nEnd of demo')

```
