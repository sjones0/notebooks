```python
# compute the collatz sequence, saving each value in a list
def collatz_sequence(x):
    # create list with the beginning value as the initial element
    num_seq = [x]
    # if result is not 1, compute next value in sequence, otherwise we're done
    while x > 1:
       # if value is even divide by 2 (using integer division "//")
       if x % 2 == 0:
         x = x // 2
       # value is odd, compute next value as 3x+1
       else:
         x = 3 * x + 1
       #append value to list
       num_seq.append(x)

    # return the list to caller
    return num_seq


print('Compute and display Collatz sequences')
while True:
    x = int(input('\nEnter a number: '))
    c = collatz_sequence(x)
    for n in c:
        print(f'{n:,}',end=' \u2022 ')
#    print(f'{c}\nThe sequence for {x} has {len(c):,} items, the max value is {max(c):,}')
    print(f'\nThe sequence for {x:,} has {len(c):,} items, the max value is {max(c):,}')


```
