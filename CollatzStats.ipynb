```python
# get stats on collatz sequences from 1 to n
# compute the collatz sequence, saving each step in a list

import time
def collatz_sequence(x):
    num_seq = [x]
    while x > 1 :
       if x % 2 == 0:
         x = x // 2
       else:
         x = 3 * x + 1
       num_seq.append(x)

    return num_seq


print('Compute the maximum sequence length and value for all sequences from 1 to n')
x = int(input('\nEnter a number for the largest sequence to compute: ').replace(',',''))

maxCount = {'n':0,'max':0}
maxValue = {'n':0,'max':0}

tic = time.perf_counter()
start = tic

for i in range(1,x+1):
    if time.perf_counter() - tic >= 1:              # more than 1 second elapsed?
        print('.',sep = '',end = '',flush=True)     # print a dot
        tic = time.perf_counter()
    c = collatz_sequence(i)

    if len(c) > maxCount['max']:                    # update max count
        maxCount['max'] = len(c)
        maxCount['n'] = i

    if max(c) > maxValue['max']:                    # update max value
        maxValue['max'] = max(c)
        maxValue['n'] = i

print(f' {time.perf_counter()-start:.2f} sec.')     # pritn elapsed time
print(f'Maximum item count for all sequences from 1 to {x:,}: {maxCount["max"]:,} for sequence {maxCount["n"]:,}')
print(f'Maximum value found in all sequences from 1 to {x:,}: {maxValue["max"]:,} in sequence {maxValue["n"]:,}')
```
